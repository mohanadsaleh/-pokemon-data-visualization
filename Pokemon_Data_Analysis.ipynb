```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

# Veri kümesini yükle

```python
veri = pd.read_csv("Pokemon.csv", index_col=0)
veri.head(10)
```

# Attack ve Defense arasındaki ilişki

```python
sns.lmplot(x='Attack', y='Defense', data=veri)
```

# Regresyon çizgisi olmadan, evreye göre renklendirme

```python
sns.lmplot(x='Attack', y='Defense', fit_reg=False, hue='Stage', data=veri)
plt.ylim(0, None)
plt.xlim(0, None)
```

# Kutu grafiği - tüm sütunlar için

```python
sns.boxplot(data=veri)
plt.xticks(rotation=45)
```

# Belirli sütunlar için kutu grafiği

```python
veri_gecici = veri.drop(['Total', 'Stage', 'Legendary'], axis=1)
sns.boxplot(data=veri_gecici, notch=True, linewidth=2.5, width=0.50)
```

# Keman grafiği

```python
sns.violinplot(x="Type 1", y="Attack", data=veri)
plt.xticks(rotation=45)
```

# Serpme (swarm) grafiği

```python
sns.swarmplot(x="Type 1", y="HP", data=veri)
plt.xticks(rotation=45)
```

# Keman ve swarm grafiğini birleştirme

```python

sns.violinplot(x="Type 1", y="Attack", data=veri, inner=None)
sns.swarmplot(x="Type 1", y="Attack", data=veri, color="red", alpha=0.7)
plt.title("Attack by Type")
plt.xticks(rotation=40)
```

# Gözlemleme

```python
veri_gecici.head()
```

# Geniş veri formatını dar formata çevirme (melt)

```python
melted_df = pd.melt(veri_gecici, id_vars=["Name", "Type 1", "Type 2"], var_name="Stat")
melted_df.head(5)
```

# Stat değişkenine göre swarmplot

```python
sns.swarmplot(x="Stat", y="value", data=melted_df, hue="Type 1")
```

# Genişlik ve sınır ayarları ile swarmplot

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x="Stat", y="value", data=melted_df, hue="Type 1", dodge=True)
plt.ylim(0, 260)
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

# Korelasyon matrisi ve ısı haritası

```python
sayisal_veri = veri_gecici.select_dtypes(include=[float, int])
corr = sayisal_veri.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

# Dağılım grafiği (histogram)

```python
sns.distplot(veri.Attack)
```

# Sütun grafiği: kategorik değişkenler

```python
sns.countplot(x="Type 1", data=veri)
plt.xticks(rotation=45)
```

# Faktör grafiği

```python

g = sns.catplot(
    x="Type 1",
    y="Attack",
    data=veri,
    hue="Stage",
    col="Stage",
    kind="swarm",
    height=5,
    aspect=1.5
)
g.set_xticklabels(rotation=-45)
plt.show()
```

# Yoğunluk grafiği

```python
sns.kdeplot(
    x=veri['Attack'],
    y=veri['Defense'],
    fill=True,
    cmap="Blues",
    thresh=0.05,
)
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.show()
```

# Ortak dağılım grafiği

```python
sns.jointplot(x="Attack", y="Defense", data=veri)
```

# Soru 1: Pokemon türlerinin dağılımı

```python

sns.countplot(
    x="Type 1",
    data=veri,
    palette="viridis",
    order=veri['Type 1'].value_counts().index
)
sns.despine()
plt.title("Pokemon Türlerinin Dağılımı")
plt.xlabel("Tür (Type 1)")
plt.ylabel("Frekans")
plt.xticks(rotation=45)
```

# Soru 2: Type ve Total arasındaki ilişki

```python

sns.boxplot(x="Type 1", y="Total", data=veri, palette="coolwarm")
sns.despine()
plt.title("Type 1 ve Total İlişkisi")
plt.xlabel("Tür (Type 1)")
plt.ylabel("Toplam İstatistik (Total)")
plt.xticks(rotation=45)
plt.show()
```

# Soru 3: Attack ve Defense arasındaki ilişki

```python

sns.scatterplot(x="Attack", y="Defense", data=veri, hue="Legendary", palette="deep", s=100)
sns.despine()
plt.title("Attack ve Defense Arasındaki İlişki")
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.legend(title="Legendary", loc="upper right") 
plt.show()
```

# Soru 4: Efsanevi ve normal Pokemon'ların toplam istatistikleri

```python
sns.violinplot(x="Legendary", y="Total", data=veri, palette="muted")
sns.despine()
plt.title("Efsanevi ve Normal Pokemonların Toplam İstatistikleri")
plt.xlabel("Legendary (Efsanevi)")
plt.ylabel("Toplam İstatistik (Total)")
plt.xticks([0, 1], ["Hayır", "Evet"])
```
